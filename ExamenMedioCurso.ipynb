```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 2. Carga de datos desde la URL oficial de Seaborn
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv"
df = pd.read_csv(url)

# 3. Mostrar las últimas 5 filas y resumen descriptivo
print("--- Últimas 5 filas del dataset ---")
display(df.tail(5))

print("\n--- Resumen estadístico de columnas numéricas ---")
display(df.describe())

# 4. Limpieza: Eliminar filas donde x, y o z sean 0
# Buscamos las filas con dimensiones nulas y las eliminamos
filas_antes = df.shape[0]
df = df[(df['x'] > 0) & (df['y'] > 0) & (df['z'] > 0)]
filas_despues = df.shape[0]
print(f"\nSe eliminaron {filas_antes - filas_despues} filas con dimensiones en cero.")

# 5. Creación de la columna volumen_aprox
df['volumen_aprox'] = df['x'] * df['y'] * df['z']

print("\nColumna 'volumen_aprox' creada con éxito.")
```

```python
# Configuramos un estilo visual más moderno
plt.style.use('seaborn-v0_8-whitegrid')

# 1. Gráfico de Caja (Boxplot) del Precio
plt.figure(figsize=(8, 6))
plt.boxplot(df['price'], patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='blue'),
            medianprops=dict(color='red', linewidth=2))
plt.title('Distribución de Precios (Identificación de Outliers)', fontsize=14, fontweight='bold')
plt.ylabel('Precio (USD)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 2. Gráfico de Dispersión (Scatter): Volumen vs Precio
plt.figure(figsize=(10, 6))
plt.scatter(df['volumen_aprox'], df['price'], alpha=0.4, c='teal', edgecolors='w', s=20)
plt.title('Relación entre Volumen y Precio', fontsize=14, fontweight='bold')
plt.xlabel('Volumen Aproximado ($x * y * z$)', fontsize=12)
plt.ylabel('Precio (USD)', fontsize=12)
plt.tight_layout()
plt.show()

# 3. Gráfico de Barras: Conteo por tipo de Corte (Cut)
plt.figure(figsize=(10, 6))
conteo = df['cut'].value_counts()
colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

conteo.plot(kind='bar', color=colores, edgecolor='black', alpha=0.8)
plt.title('Cantidad de Piezas por Tipo de Corte', fontsize=14, fontweight='bold')
plt.xlabel('Calidad del Corte', fontsize=12)
plt.ylabel('Número de Diamantes', fontsize=12)
plt.xticks(rotation=0) # Mantiene las etiquetas horizontales para que se lean mejor
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
